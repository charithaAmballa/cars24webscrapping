```python
location_counts = df.LOCATION.value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=location_counts.index, y=location_counts.values)
plt.xlabel('Number of Cars')
plt.ylabel('Location')
plt.title('Number of Cars by Location')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
fuel_counts = df.FUEL_TYPE.value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=fuel_counts.index, y=fuel_counts.values)
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by Fuel Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
transmission_counts = df.TRANSMISSION.value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=transmission_counts.index, y=transmission_counts.values)
plt.xlabel('transmission Type')
plt.ylabel('Number of Cars')
plt.title('Number of Cars by transmission Type')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
